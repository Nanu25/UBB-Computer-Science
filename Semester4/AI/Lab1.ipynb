```python

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(44)

t = np.arange(0, 30, 0.01)
nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2

# Two signals with a coherent part at 10 Hz and a random part
s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, NFFT=256, Fs=1. / 0.01)
axs[1].set_ylabel('Coherence')

plt.show()
```

```python
import numpy as np
#more elegante way.
def spinner(probabilityDistribution):
    random_number = np.random.random()
    
    cumulative_probability = 0.0
    for index, probability in enumerate(probabilityDistribution):
        cumulative_probability += probability
        if random_number < cumulative_probability:
            return index


probabilityDistribution = [0.1, 0.2, 0.3, 0.4] #every probability has its own interval. 
print(spinner(probabilityDistribution))
```

```python
import numpy as np

def spinner_alternative(probability_distribution):
    probabilities = np.array(probability_distribution)
    indices = np.arange(len(probabilities))
    return np.random.choice(indices, p=probabilities)

probability_distribution = [0.1, 0.2, 0.3, 0.4]
print(spinner_alternative(probability_distribution))
```

```python
#6B.
import numpy as np


def angle_spinner(angleList):
    total_angle = sum(angleList)  
    probabilityDistribution = [angle / total_angle for angle in angleList] 
    
    random_number = np.random.random()  
    cumulative_probability = 0.0
    
    for index, probability in enumerate(probabilityDistribution):
        cumulative_probability += probability
        if random_number < cumulative_probability:
            return index  

angles = [30, 60, 90, 180]  
print(angle_spinner(angles))
```

```python
#7
import numpy as np

def calculate_ball_in_pipe_probability(num_simulations=100000):
    R = 30  
    r = 1   
    l = 100  
    
    x = np.random.uniform(-l/2, l/2, num_simulations)
    y = np.random.uniform(-l/2, l/2, num_simulations)
    
    distances = np.sqrt(x**2 + y**2)
    inside_pipe = distances < (R - r)
    probability_simulated = np.mean(inside_pipe)
    probability_theoretical = np.pi * (R - r)**2 / (l**2)

    return probability_simulated, probability_theoretical

simulated_prob, theoretical_prob = calculate_ball_in_pipe_probability()
print(f"Simulated probability: {simulated_prob:.6f}")
print(f"Theoretical probability: {theoretical_prob:.6f}")
print(f"Difference: {abs(simulated_prob - theoretical_prob):.6f}")
```
